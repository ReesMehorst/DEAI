```python
def optellen(getal1, getal2):
    return getal1 + getal2

def aftrekken(getal1, getal2):
    return getal1 - getal2

def vermenigvuldigen(getal1, getal2):
    return getal1 * getal2

def delen(getal1, getal2):
    return getal1 / getal2

getal1 = float(input("Kies een getal:"))
getal2 = float(input("Kies nog een getal:"))
methode = int(input("1. optellen 2. aftrekken 3. vermenigvuldigen 4. delen. Kies 1-4:"))
match methode:
    case 1:
        print(optellen(getal1, getal2))
    case 2:
        print(aftrekken(getal1, getal2))
    case 3:
        print(vermenigvuldigen(getal1, getal2))
    case 4:
        print(delen(getal1, getal2))
    case _: #default case
        print("verkeerde input")
```
